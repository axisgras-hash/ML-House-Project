# ML Housing Price Prediction project

```python
# Step 1: Load Important Modules
import os
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
print("All Modules Loaded Successfully!!")
```

```python
# Step 2: Load Datasets
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()  # Required Internet
print("Data Loaded Successfully!!")
```

```python
# Step 3: Convert data to DF
df = pd.DataFrame(data['data'], columns = data['feature_names'])
df['target'] = data['target']
df.sample(5)
```

```python
cols = ["MedInc"	,"HouseAge"	,"AveRooms"	,"AveBedrms"	,"Population"	,"AveOccup"]
print(cols)
```

```python
# Step 4: Divide dataset into X and y
# without lat | long
X = df[cols]
X
```

```python
y = df['target']
y
```

```python
# Divide train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    random_state=42,
                                                    test_size = 0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Step 5: MultiModel building
from sklearn.metrics import (r2_score,
                             mean_squared_error,
                             mean_absolute_error,
                             root_mean_squared_error)
```

```python
all_models_name = ['Linear Regression',
                  'Decison Tree','Random Forest',
                  'Knn','SVM']

all_models = [LinearRegression,DecisionTreeRegressor,
              RandomForestRegressor,
              KNeighborsRegressor,SVR]

comp_df = {
           'Model_name':[],
           'Model_R2_Score':[],
           'Random State':[],
           'MSE':[],'MAE':[],'RMSE':[]}

for r in range(1,50):
  for index,value in enumerate(all_models_name):
    X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    random_state=r,
                                                    test_size = 0.2)
    print(f"Model Training for {r}: {value}")
    model = all_models[index]()

    # Training
    model.fit(X_train,y_train)

    # Model Score R2

    y_pred = model.predict(X_test)
    score = r2_score(y_test,y_pred)

    mae = mean_absolute_error(y_test,y_pred)
    mse = mean_squared_error(y_test,y_pred)
    rmse = root_mean_squared_error(y_test,y_pred)


    #===========Store in dict=============
    comp_df['Model_name'].append(value)
    comp_df['Model_R2_Score'].append(score)
    comp_df['Random State'].append(r)
    comp_df['MSE'].append(mse)
    comp_df['MAE'].append(mae)
    comp_df['RMSE'].append(rmse)

  # time.sleep(2)
    display(clear=True)

print("Done")
```

```python
model_comparison_df = pd.DataFrame(comp_df)
model_comparison_df
```

```python
temp_data = model_comparison_df.groupby('Model_name')['Model_R2_Score'].max()
graph_x = temp_data.index
graph_y = temp_data.values


plt.figure(figsize = (15,8))
plt.title("Multi Model Comparison")
ax = sns.barplot(x = graph_x, y= graph_y, palette = sns.color_palette('rainbow'))
# for i in ax.containers:
#   plt.bar_label(ax)
plt.xlabel("ML Models")
plt.ylabel("Model Score")
plt.show()
```

```python
model_comparison_df[model_comparison_df['Model_name'] == 'Random Forest'].sort_values(by = 'Model_R2_Score', ascending = False)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    random_state=44,
                                                    test_size = 0.2)
final_model = RandomForestRegressor()

final_model.fit(X_train,y_train)
final_model.score(X_test,y_test)
```

```python
# Model Save
import pickle
with open("house_price_gpt.pkl", 'wb') as f:
  pickle.dump(final_model,f)
print("done")
```
